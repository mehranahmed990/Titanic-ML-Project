```python
# I want to add a new column gender, which can create a new level child.
# If age is less than 18, (s)he is a child.
def mfc(rahul):
    age, sex = rahul

    if age < 18:
        return 'child'
    else:
        return sex
```

## Unsupervised Learning (KMeans Clustering)


```python
# package required for data and Kmeans
from sklearn import datasets
from sklearn.cluster import KMeans
import sklearn.metrics as sm
#from scipy.spatial.distance import cdist
# packages required for preprocessing of the data
#from sklearn.preprocessing import LabelEncoder
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.info()
```

```python
iris.species.value_counts()
```

```python
iris1 = iris.drop('species', axis = 1)
```

```python
iris1km = KMeans(n_clusters = 3, random_state = 10)
```

```python
iris1km.fit(iris1)
```

```python
# describe this code output
iris1km.cluster_centers_
```

```python
# describe the result of inertia
iris1km.inertia_

#Sure! Inertia is a measure of how well a set of data points are clustered around their centroids in a clustering algorithm.
#It calculates the sum of squared distances between each data point and the centroid of its cluster. Essentially, it gives an idea of how compact the clusters are.
#lower inertia value indicates more compact and well-defined clusters, while a higher inertia value suggests that the clusters are less well-defined and the data points are more spread out
```

```python
specieslabels = iris1km.fit_predict(iris1)
print(specieslabels)

#trains a KMeans clustering model on the Iris dataset, assigns each data point to a cluster, and then prints the assigned cluster labels.
#It's essentially performing unsupervised learning to group similar Iris flowers based on their features.
```

```python
ir = iris
```

```python
ir['pred'] = specieslabels
```

```python
ir.head()
```

```python
# Comments on these two graphs
sns.FacetGrid(ir, hue = "pred", height = 5).map(plt.scatter, "sepal_length", "sepal_width").add_legend()

#Orange points (pred = 1) cluster around sepal length 4.5-6.0 and sepal width 2.5-4.5.
#Green points (pred = 2) cluster around sepal length 5.0-7.0 and sepal width 2.0-3.5.
#Blue points (pred = 0) are more spread out but mainly around sepal length 5.0-8.0 and sepal width 2.0-3.5.
```

```python
sns.FacetGrid(ir, hue = "species", height = 5).map(plt.scatter, "sepal_length", "sepal_width").add_legend()

#Setosa (blue) generally has higher sepal widths and shorter sepal lengths.
#Versicolor (orange) has a moderate range for both sepal length and sepal width.
#Virginica (green) has the longest sepal lengths and varied sepal widths.
```

```python
# You need to change the encoding based on your output
numspecies = {'species':{'setosa':0,'versicolor':1,'virginica':2}}
numspecies
```

```python
# we will use the replace function
iris = iris.replace(numspecies)
iris.head()
```

```python
print(iris['species'][0:5])
print(iris['species'][80:90])
print(iris['species'][100:105])
```

```python
# comments on the performance metrics
sm.accuracy_score(iris.species,iris1km.labels_)

#This is quite low and suggests that the model is not performing well.
```

```python
# How many cases has been misclassfied
#confusion matrix
sm.confusion_matrix(iris.species,iris1km.labels_)

#The first row (Setosa) has 0 correct classifications and 50 misclassifications (all predicted as Versicolor).
#The second row (Versicolor) has 3 correct classifications and 47 misclassifications (all predicted as Virginica).
#The third row (Virginica) has 36 correct classifications and 14 misclassifications (mainly predicted as Setosa).
```

```python
# find out the percentage of the misclassfied cases
sm.accuracy_score(iris.species,iris1km.labels_)
```

```python
# Scree Plot
# Comments on the optimum number of cluster
wss = []
abhishek = range(1,16)
for k in abhishek:
    ethankmodel = KMeans(n_clusters=k).fit(iris1)
    wss.append(ethankmodel.inertia_)

# Plot the elbow
plt.figure(figsize=(8,6))
plt.plot(abhishek, wss, 'bx-')
plt.xlabel('k')
plt.ylabel('Within Sum of Squares')
plt.title('The Elbow Method showing the optimal k')
plt.show()

#The WSS decreases sharply from k=1 to k=3, forming an "elbow" at k=3.
#Beyond k=3, the reduction in WSS becomes marginal, indicating that adding more clusters doesn't significantly improve the clustering.
#Therefore, the optimal number of clusters is 3, as this is where the "elbow" is formed, suggesting a good balance between cluster compactness and the number of clusters.
```
